```python
def percent_visual(column):
    p=data.groupby(column)["Churn"].value_counts().to_frame()
    p=p.rename({"count":"NOC"},axis=1).reset_index()
    p["Percentage"]=p["NOC"]/p["NOC"].sum()*100
    sns.barplot(x=column,y="Percentage",hue="Churn",data=p)
```

```python
fig, axis=plt.subplots(1, 2, figsize=(12,4))
axis[0].set_title("Has partners")
axis[1].set_title("Has dependents")
y_axis="Percentage"
partner=data.groupby("Partner")["Churn"].value_counts()/len(data)
partner=partner.to_frame().rename({"count":y_axis},axis=1).reset_index()
sns.barplot(x='Partner',y=y_axis,hue="Churn",data=partner,ax=axis[0])
dependents=data.groupby("Dependents")["Churn"].value_counts()/len(data)
dependents=dependents.to_frame().rename({"count":y_axis},axis=1).reset_index()
sns.barplot(x='Dependents',y=y_axis,hue="Churn",data=dependents,ax=axis[1])
```

```python
X=data.drop(["Churn"],axis=1)
y=data['Churn']
```

```python
data.head()
```

#feature Scaling - Normalizing the numerical features 

```python
from sklearn.preprocessing import MinMaxScaler
mmscale=MinMaxScaler()
X["tenure"]=mmscale.fit_transform(X[["tenure"]])
X["TotalCharges"]=mmscale.fit_transform(X[["TotalCharges"]])
X["MonthlyCharges"]=mmscale.fit_transform(X[["MonthlyCharges"]])
```

```python
X.head()
```

```python
X[categorical_data].head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
for c in categorical_data:
    encoder.fit(X[c])
    X[c]=encoder.transform(X[c])
    mapping=dict(zip(encoder.classes_,encoder.transform(encoder.classes_)))
    print(c,":",mapping)
```

```python
y=encoder.fit_transform(y)
mapping=dict(zip(encoder.classes_,encoder.transform(encoder.classes_)))
```

```python
print(mapping)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,stratify=y, random_state=42)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
from sklearn.feature_selection import SelectKBest,f_classif
```

```python
select2=SelectKBest(f_classif,k=6)
select2.fit(X_train,y_train)
```

```python
X_train.columns[select2.get_support()]
```

```python
X_train_selected=X_train[X_train.columns[select2.get_support()]]
```

```python
X_test_selected=X_test[X_train.columns[select2.get_support()]]
```

#modeling

```python
from sklearn.tree import DecisionTreeClassifier
model1=DecisionTreeClassifier()
model1.fit(X_train_selected,y_train)
ans1=model1.predict(X_test_selected)
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,ans1)
score*100
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,ans1)
sns.heatmap(cm,annot=True,fmt='g')
```

```python
from sklearn.ensemble import RandomForestClassifier
model2=RandomForestClassifier()
model2.fit(X_train_selected,y_train)
ans2=model2.predict(X_test_selected)
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,ans2)
score*100
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,ans2)
sns.heatmap(cm,annot=True,fmt='g')
```

```python
from sklearn.svm import SVC
model3=SVC()
model3.fit(X_train_selected,y_train)
ans3=model3.predict(X_test_selected)
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,ans3)
score*100
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,ans3)
sns.heatmap(cm,annot=True,fmt='g')
```

```python
from xgboost import XGBClassifier
model4=XGBClassifier()
model4.fit(X_train_selected,y_train)
ans4=model4.predict(X_test_selected)
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,ans4)
score*100
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,ans4)
sns.heatmap(cm,annot=True,fmt='g')
```
